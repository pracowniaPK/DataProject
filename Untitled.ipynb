```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('raw.csv', decimal=',')
raw
```

```python
raw.iloc[1:,1].describe()
```

```python
def unitaryzacja(srs):
    mini = srs.describe()['min']
    rozstep = srs.describe()['max'] - mini
    return srs.map(lambda p: (p - mini)/rozstep)
 
uni = raw.iloc[1:,1:].apply(unitaryzacja)
sns.heatmap(uni, xticklabels=3)
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(15,15))
plt.title('Korelacja', y=1.05, size=15)
sns.heatmap(uni.astype(float).corr(),linewidths=0.05, vmax=1.0, 
            square=True, cmap=colormap, xticklabels=3, yticklabels=3, annot=True)
```
